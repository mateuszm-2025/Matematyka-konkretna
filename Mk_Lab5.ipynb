```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)

def swish_gradient(x):
    s = sigmoid(x)
    return s + x * s * (1 - s)

x = np.linspace(-10, 10, 1000)

y_swish = swish(x)
y_gradient = swish_gradient(x)

punkty = np.array([-10, -5, -2, -1, 0, 1, 2, 5, 10])

tabela = pd.DataFrame({
    "x": punkty,
    "sigmoid(x)": sigmoid(punkty),
    "Swish(x)": swish(punkty),
    "Swish'(x)": swish_gradient(punkty)
})

display(tabela.round(4))

plt.figure(figsize=(10, 6))

plt.plot(x, y_swish, label="Swish(x)")
plt.plot(x, y_gradient, label="Swish'(x)")

plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)

plt.title("Funkcja Swish oraz jej gradient")
plt.xlabel("x")
plt.ylabel("wartość")
plt.grid(True)
plt.legend()
plt.show()
```
