```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams['figure.figsize'] = [16,8]
```

```python
A = imread('12.webp')
X = np.mean(A, -1)

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.title('Obraz w skali szarości')
plt.show()

print("Rozmiar macierzy obrazu:", X.shape)
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)

print("Liczba wartości singularnych:", S.shape[0])

S = np.diag(S)
```

```python
# Macierz korelacji wierszy
XXT = U @ S @ S @ U.T

# Macierz korelacji kolumn
XTX = VT.T @ S @ S @ VT

print("Rozmiar macierzy XX^T:", XXT.shape)
print("Rozmiar macierzy X^T X:", XTX.shape)
```

```python
plt.figure(figsize=(8,8))
img = plt.imshow(XXT)
img.set_cmap('gray')
plt.axis('off')
plt.title('Macierz korelacji wierszy: XX^T')
plt.show()
```

```python
plt.figure(figsize=(8,8))
img = plt.imshow(XTX)
img.set_cmap('gray')
plt.axis('off')
plt.title('Macierz korelacji kolumn: X^T X')
plt.show()
```

```python
plt.figure()
plt.semilogy(np.diag(S))
plt.title('Wartości singularne')
plt.show()
```
